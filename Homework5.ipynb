```python
#W5A16

def armstrong(x):
    if x<0:
        return False

    mem = x
    sum = 0
    i = len(str(x))
    
    while x>0:
        sum = sum + (x%10)**i
        x = x//10

    return sum == mem

n = int(input())

print(armstrong(n))
```

```python
#W5A20

def number_type(s):
    
    if s.isdigit():
        return 'int'
    
    if '.' in s:
        char = s.split('.')
        if len(char)==2 and (char[0].isdigit() and char[1].isdigit()):
            return 'float'
        
    return 'neither'

s = str(input())

print(number_type(s))
```

```python
#W5A22

def sumlr(s):
    sum = 0
    l = 0
    for i in range(0,6):
        sum = sum + int(s[i])
        if i<3:
            l = l + int(s[i])

    return [l,sum-l]

s = str(input())

print(sumlr(s))
```

```python
#W5A29

def max(a,b,c):
    if a > b:
        if b > c:
            return [a,c]
        else:
            if a > c:
                return [a,b]
            else:
                return [c,b]
    else:
        if a > c:
            return [b,c]
        else:
            if b > c:
                return [b,a]
            else:
                return [c,a]
            
a,b,c = map(int,input().split())

print(max(a,b,c))
```

```python
#W5A30

def c(n,k):
    if k > n - k:
        k = n - k
    
    m = 1
    for i in range(0,k):
        m = m*(n-i)
        m = m/(i+1)

    return m

n, k = map(int,input().split())

print(c(n,k))
```

```python
#W5A33
#Giong bai #1 file ontap1

def smart(type, hour, km, wait):
    total = 0

    if type==4:
        if km<=0.8:
            total = 11000
        else:
            if km<=20:
                total = 11000 + 13000*(20-0.8)
            else:
                total = 11000 + 13000*19.2 + 10500*(km-20)
    else:
        if km<=0.8:
            total = 12000
        else:
            if km<=20:
                total = 12000 + 14000*(20-0.8)
            else:
                total = 12000 + 14000*19.2 + 12500*(km-20)

    if hour in [22,23,0,1,2,3,4,5]:
        total = total*1.2

    if wait>5:
        total = total + int(wait+1)*2000

    if km>30 and not (hour in [22,23,0,1,2,3,4,5]):
        total = total*0.9

    if km>5 and total<50000:
        total = 50000

    return total
```

```python
#W5A34

def dm(x):
    s = []
    while x>0:
        mem = x%10
        if mem in s:
            return False
        s.append(mem)
        x = x//10
    return True

def sum_1_n(x):
    sum = 0
    for i in range(1,x+1):
        if dm(i):
            sum = sum + i
    return sum

n = int(input())

print(sum_1_n(n))
```
