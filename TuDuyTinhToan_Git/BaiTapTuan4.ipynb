```python
#W4A8

def re(x):
    return int(str(x)[::-1])

def dx(x):
    return str(x) == str(x)[::-1]

n = int(input())
step = 0

while not dx(n):
    step = step + 1
    n = n + re(n)

print(f'{step} {n}')
```

```python
#W4A9

def dm(x):
    s = []
    while x > 0:
        u = x%10
        if u in s:
            return False
        s.append(u)
        x = x//10
    return True

n = int(input())

for i in range(1, int(n**(1/2))+1):
    if dm(i*i):
        print(i*i, end=' ')
```

```python
#W4A10

def collatz(x):
    n = 1
    while x!=1:
        if x%2==0:
            x = x/2
            
            n = n+1
        else:
            x = x*3+1
            
            n = n+1
        if x==1:
            return n
    return n

n = int(input())
l = 0

for i in range(1,n+1):
    if collatz(i)>l:
        l = collatz(i)
        x = i

print(x, l)
```

```python
#W4A11

n = int(input())

def uc(x):
    a = 0
    while x%2==0:
        a = a + 1
        x = x/2
    if a == 0:
        return 0
    
    b = 1
    i = 3
    while i*i <= x:
        count = 0
        while x%i == 0:
            count = count + 1
            x = x/i
        if count > 0:
            b = b*(count+1)
        i = i + 2
    
    if x>1:
        b = b*2
    return a*b

print(uc(n))
```

```python
#W4A13

def sumu(x):
    sum = 1
    for i in range(2, int(x**(1/2))+1):
        if x%i == 0:
            sum = sum + i
            if i != (x//i):
                sum = sum + x//i
    return sum

a, b = map(int, input().split())

if a == sumu(b) and b == sumu(a):
    print('true')
else:
    print('false')
```

```python
#W4A14

m, n = map(int, input().split())

def ucln(m,n):
    if m == 0 or n == 0:
        return m + n
    while m != n:
        if m > n:
            m = m - n
        else:
            n = n - m
    return m

print(ucln(m,n))
```
