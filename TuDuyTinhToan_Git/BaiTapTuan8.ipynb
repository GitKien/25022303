```python
#W8A16

arr = list(map(int, input().split()))
dic = {}

for i in arr:
    if i in dic:
        dic[i] += 1
    else:
        dic[i] = 1

y = 0
for i in dic:
    if i>dic[i]:
        y += dic[i]
    elif i<dic[i]:
        y += dic[i]-i

if y==0:
    print('YES')
else:
    print('NO', y)
```

```python
#W8A27

def ktr(x):
    arr = []
    for i in x:
        if i in '{[(<':
            arr.append(i)
        else:
            if not arr:
                return 'NO'
            
            mem = arr.pop()
            if (i=='}' and mem!='{') or (i==']' and mem!='[') or (i==')' and mem!='(') or (i=='>' and mem!='<'):
                return 'NO'
    
    if not arr:
        return 'YES'
    else:
        return 'NO'

s = str(input())
print(ktr(s))
```

```python
#W8A28

def hs(x):
    if x<1:
        return False
    for i in range(2,x):
        if x%i==0:
            return True
    return False

n = int(input())
max = 0
out = ''

i = 0
while i<n:
    mem = ''
    count = 0
    while i<n and hs(i):
        mem = mem + str(i) + ', '
        i += 1
        count += 1
    if count> max:
        max = count
        out = mem
    i += 1

print(out[:-2])
```

```python
#W8A29

arr = list(map(int, input().split()))
max = 0
x = 0

i = 1
while i<len(arr):
    count = 1
    mem = i-1
    while i<len(arr) and arr[i-1]<arr[i]:
        count += 1
        i += 1
    if count>max:
        max = count
        x = mem
    i += 1

for i in range(x,x+max):
    print(arr[i], end=' ')
```

```python
#W8A30

def trh(x):
    if dp[x]!=-1:
        trh(dp[x])
    print(arr[x], end=' ')

arr = list(map(int, input().split()))
dp = [-1]*len(arr)
mem = [1]*len(arr)

k = 1
for i in range(len(arr)):
    for j in range(i):
        if arr[i]>arr[j] and mem[j]+1>mem[i]:
            mem[i] = mem[j] + 1
            dp[i] = j
            if mem[i]>k:
                k = mem[i]

for i in range(len(arr)-1,-1,-1):
    if mem[i]==k:
        trh(i)
        break
```
