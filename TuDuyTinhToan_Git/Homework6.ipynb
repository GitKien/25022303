```python
#W6A16

n, m = map(int, input().split())
a = []

for i in range(n):
    mem = list(map(int, input().split()))
    a.append(mem)

for i in range(n):
    for j in range(m):
        print(f'{a[i][j]:4}', end=' ')
    print()
```

```python
#W6A18

n, m, k = map(int, input().split())
sum = 0
a = []

for i in range(n):
    mem = list(map(int, input().split()))
    a.append(mem)

if k>0 and k<=m:
    for i in range(n):
        sum = sum + a[i][k-1]

print(sum)
```

```python
#W6A21

a = list(map(str, input().split()))
dic = {}

for i in range(len(a)):
    mem = a[i]
    if mem not in dic:
        dic[mem] = []
    dic[mem].append(i+1)

for i in dic:
    dic[i] = tuple(dic[i])

print(dic)
```

```python
#W6A23

a = list(map(int, input().split()))
x = int(input())

i = 0
while i<len(a)-1:
    if a[i]<x:
        del a[i]
    else:
        i = i + 1

print(a)
```

```python
#W6A26

a = list(map(int, input().split()))
sum = 0

for i in range(1,len(a)-1):
    mem = a[i]
    if mem>a[i-1] and mem>a[i+1]:
        sum = sum + mem

print(sum)
```
