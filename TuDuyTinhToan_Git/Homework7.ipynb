```python
#W7A27

s = str(input())

def ktr(s):
    arr = []
    for i in s:
        if i in '{[(':
            arr.append(i)
        else:
            if not arr:
                return 'Khong hop le'
        
            mem = arr.pop()
            if (i == ')' and mem != '(') or (i == ']' and mem != '[') or (i == '}' and mem != '{'):
                return 'Khong hop le'

    if not arr:
        return 'Hop le'
    else:
        return 'Khong Hop le'
    
ktr(s)
```

```python
#W7A28

def hs(x):
    if x<1:
        return False
    for i in range(2,x):
        if x%i == 0:
            return True
    return False

def ngto(x):
    if x<2:
        return False
    for i in range(2,int(x**(1/2))+1):
        if x%i==0:
            return False
    return True

n = int(input())
max = 0
mem = 0

for i in range(n):
    if hs(i) and not ngto(i):
        mem += 1
    else:
        if mem>max:
            max = mem
        else:
            mem = 0
```
