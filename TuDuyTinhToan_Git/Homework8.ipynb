```python
#W8A16

class shape:
    def area(self):
        return 0
    
class rectangle(shape):
    def __init__(self, w, h):
        self.w = w
        self.h = h
    def area(self):
        return self.h * self.w
    
class circle(shape):
    def __init__(self, r):
        self.r = r
    def area(self):
        return 3.14 * self.r * self.r
    
s = str(input())

if s == 'rectangle':
    w, h = map(float, input().split())
    m = rectangle(w, h)
else:
    r = float(input())
    m = circle(r)

print(f'{m.area():.2f}')
```

```python
#W8A17

class vehicle:
    def __init__(self, brand, speed):
        self.brand = brand
        self.speed = speed
    def info(self):
        return f'{self.brand}-{self.speed}'

class car(vehicle):
    def __init__(self, brand, speed, seats):
        super().__init__(brand, speed)
        self.seats = seats
    def info(self):
        return f'{self.brand}-{self.speed}-{self.seats}'
    
s = str(input())
if s == 'vehicle':
    brand, speed = input().split()
    m = vehicle(brand, speed)
elif s == 'car':
    brand, speed, seats = input().split()
    m = car(brand, speed, seats)

print(m.info())
```

```python
#W8A18

class polygon:
    def perimeter():
        return 0
    
class triangle(polygon):
    def __init__(self, a, b, c):
        self.a = a
        self.b = b
        self.c = c
    def perimeter(self):
        return self.a + self.b + self.c
    
a, b, c = map(int, input().split())
m = triangle(a, b, c)
print(m.perimeter())
```

```python
#W8A19

class person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    def describle(self):
        return f'{self.name} {self.age}'
    
class student(person):
    def __init__(self, name, age, school):
        super().__init__(name, age)
        self.school = school
    def describle(self):
        return f'{self.name} {self.age} {self.school}'
    
s = str(input())
if s == 'person':
    name, age = input().split()
    m = person(name, age)
elif s == 'student':
    name, age, school = input().split()
    m = student(name, age, school)

print(m.describle())
```

```python
#W8A20

class account:
    def __init__(self, balance):
        self.balance = balance
    def deposit(self, x):
        self.balance += x
    def get_balance(self):
        return self.balance
    
class savingaccount(account):
    def __init__(self, balance, rate):
        super().__init__(balance)
        self.rate = rate
    def add_interest(self):
        self.balance += self.balance * self.rate

balance, rate = map(float, input().split())

acc = savingaccount(balance, rate)
acc.add_interest()

print(f'{acc.get_balance():.2f}')
```

```python
#W8A21

class printer:
    def print_value(self, x):
        print(x)

class binaryprinter(printer):
    def print_value(self, x):
        binary = ''

        while x>0:
            binary = str(x%2) + binary
            x = x//2

        print(binary)

n = int(input())
m = binaryprinter()

m.print_value(n)
```

```python
#W8A22

class matrix:
    def __init__(self, a, b, c, d):
        self.a = a
        self.b = b
        self.c = c
        self.d = d
    def trace(self):
        return self.a + self.d
    def determinant(self):
        return self.a * self.d - self.b * self.c
    
a, b, c, d = map(int,input().split())
m = matrix(a, b, c, d)
print(m.trace(), m.determinant())
```

```python
#W8A23

class polynomial:
    def __init__(self, a, b):
        self.a = a
        self.b = b
    def evaluate(self, x):
        return self.a * x + self.b
    
a, b, x = map(int, input().split())
m = polynomial(a,b)
print(m.evaluate(x))
```

```python
#W8A24

class polynomial2:
    def __init__(self, a, b, c):
        self.a = a
        self.b = b
        self.c = c
    def evaluate(self, x):
        return self.a * x * x + self.b *x + self.c
    
a, b, c, x = map(int, input().split())
m = polynomial2(a, b, c)
print(m.evaluate(x))
```

```python
#W8A25

class polynomialsum:
    def __init__(self, a, b, c, d):
        self.a = a
        self.b = b
        self.c = c
        self.d = d
    def sum_str(self):
        k = self.a + self.c
        m = self.b + self.d
        if m>0:
            return f'{k}x + {m}'
        else:
            return f'{k}x - {-m}'
        
a, b, c, d = map(int, input().split())
m = polynomialsum(a, b, c, d)
print(m.sum_str())
```

```python
#W8A27

class product:
    def __init__(self, name, price):
        self.name = name
        self.price = price

class discountproduct(product):
    def __init__(self, name, price, percent):
        super().__init__(name, price)
        self.percent = percent
    def final_price(self):
        return self.price * (100 - self.percent) / 100
    
name, price, percent = input().split()
m = discountproduct(name, int(price), int(percent))
print(f'{m.final_price():.2f}')
```

```python
#W8A28

class counter:
    def __init__(self):
        self.count = 0
    def inc(self):
        self.count += 1
    def dec(self):
        self.count -= 1
    def value(self):
        return self.count
    
n = int(input())
arr = list(map(str,input().split()))

m = counter()
for i in arr:
    if i=='INC':
        m.inc()
    elif i=='DEC':
        m.dec()

print(m.value())
```

```python
#W8A29

class time:
    def __init__(self, h, m, s):
        self.h = h
        self.m = m
        self.s = s
    def to_seconds(self):
        return self.h * 3600 + self.m * 60 + self.s
    
h, m, s = map(int,input().split())
m = time(h,m,s)
print(m.to_seconds())
```

```python
#W8A30

class stack:
    def __init__(self):
        self.arr = []
    def push(self, x):
        self.arr.append(x)
    def pop(self):
        self.arr.pop()
    def top(self):
        print(self.arr[-1])

n = int(input())

m = stack()
for i in range(n):
    s = input().split()
    if s[0]=='PUSH':
        m.push(s[1])
    elif s[0]=='POP':
        m.pop()
    elif s[0]=='TOP':
        m.top()
```
