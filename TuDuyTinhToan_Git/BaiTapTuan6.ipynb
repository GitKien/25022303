```python
#W6A9

def indic():
    a = str(input())
    if a == '{}':
        return {}
    a = a.strip('{}').split(', ')
    dic = {}
    for i in range(0,len(a)):
        mem1, mem2 = map(str, a[i].split(': '))
        mem1 = mem1.strip("''")
        dic[mem1] = int(mem2)
    return dic

dic1 = indic()
dic2 = indic()

for i in dic2:
    if i in dic1:
        dic1[i] = dic1[i] + dic2[i]
    else:
        dic1[i] = dic2[i]

print(dic1)
```

```python
#W6A10

a = list(map(int, input().split()))
k = int(input())
out =[]

for i in range(0, len(a)-1):
    mem1 = a[i]
    mem2 = k - mem1
    for j in range(i+1,len(a)):
        if a[j]==mem2:
            mem = (mem1,mem2)
            if mem not in out:
                out.append(mem)
            break

for i in range(len(out)-1):
    min = out[i][0]
    mem_min = i
    for j in range(i+1,len(out)):
        if out[j][0]<min:
            min = out[j][0]
            mem_min = j
    out[i], out[mem_min] = out[mem_min], out[i]

print(out)
```

```python
#W6A12

a = list(map(int, input().split()))
m = {}
dic = []

for i in range(len(a)):
    mem = a[i]
    if mem in m:
        m[mem] += 1
    else:
        m[mem] = 1
        dic.append(mem)

max = m[dic[0]]
out = dic[0]
for i in range(1,len(dic)):
    mem = dic[i]
    if (m[mem]>=max) and (mem<out):
        max = m[mem]
        out = mem

print(out)
```
