```python
#W7A3
#SelectionSort

arr = list(map(int,input().split()))

for i in range(len(arr)-1):
    min = arr[i]
    mem = i
    for j in range(i+1,len(arr)):
        if arr[j] <= min:
            min = arr[j]
            mem = j
    arr[i], arr[mem] = arr[mem], arr[i]

print(arr)
```

```python
#W7A3
#BubbleSort

arr = list(map(int,input().split()))

for i in range(len(arr)):
    for j in range(0,len(arr)-i-1):
        if arr[j]>arr[j+1]:
            arr[j], arr[j+1] = arr[j+1], arr[j]

print(arr)
```

```python
#W7A8
#nhambai
def se(arr,x):
    l = 0
    r = len(arr)
    while l<r:
        mid = (r+l)//2
        if arr[mid]>=x:
            r = mid
        else:
            l = mid + 1
    return l

arr = [1, 9, 7, 8, 5]
sort = []
result = []

for i in range(len(arr)-1,-1,-1):
    k = se(sort,arr[i])
    result.append(k)
    sort.insert(k,arr[i])

print(result[::-1])
```

```python
#W7A9

def se(arr,x):
    l = 0
    r = len(arr)
    while l<r:
        mid = (l+r)//2
        if mid<x:
            l = mid + 1
        else:
            r = mid
    return l

arr = list(map(int, input().split()))
sort = []
result = []

for i in range(len(arr)-1,-1,-1):
    k = se(sort,arr[i])
    result.append(k)
    sort.append(k,arr[i])

print(result[::-1])
```

```python
#W7A10

arr = list(map(int,input().split()))
arr.insert(0,0)
k = int(input())

for i in range(1,len(arr)):
    arr[i] = arr[i-1] + arr[i]

max = 0
for i in range(1,len(arr)-1):
    for j in range(i,len(arr)):
        sum = arr[j] - arr[i-1]
        if sum%k>max:
            max = sum%k

print(max)
```

```python
#W7A11

def search(s,k):
    k = k.split()
    l = list(s.split())
    
    for i in range(len(l)):
        if (not l[i][-1].isdigit()) and (not l[i][-1].isalpha()):
            l[i] = l[i][:-1]
    
    count = 0
    for i in range(len(l)-len(k)+1):
        if l[i:i+len(k)]==k:
            count += 1
    return count

s = str(input())
k = str(input())

search(s,k)
```
