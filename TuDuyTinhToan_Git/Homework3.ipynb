```python
#bai4

h = int(input())
d = 1 + 2*(h-1)

for i in range(0,h):
    s = ' '*(d//2-i)
    if i==0:
        s = s + '*'
    elif i==h-1:
        s = '*'*d
    else:
        s = s + '*' + ' '*(1+(i-1)*2) + '*'
    print(s)
```

```python
def sumu(x):
    sum = 0
    for i in range(1,int(x**(1/2))+1):
        if x%i==0:
            sum = sum + i
            if (x//i != i) and (x//i != x):
                sum = sum + x//i
    return sum

a = int(input())

print(a==sumu(a))
```
