```python
#1

def smart(type, hour, km, wait):
    total = 0

    if type==4:
        if km<=0.8:
            total = 11000
        else:
            if km<=20:
                total = 11000 + 13000*(20-0.8)
            else:
                total = 11000 + 13000*19.2 + 10500*(km-20)
    else:
        if km<=0.8:
            total = 12000
        else:
            if km<=20:
                total = 12000 + 14000*(20-0.8)
            else:
                total = 12000 + 14000*19.2 + 12500*(km-20)

    if hour in [22,23,0,1,2,3,4,5]:
        total = total*1.2

    if wait>5:
        total = total + int(wait+1)*2000

    if km>30 and not (hour in [22,23,0,1,2,3,4,5]):
        total = total*0.9

    if km>5 and total<50000:
        total = 50000

    return total
```
