```python
#W4A18

a, b = map(int, input().split())

while a!=b:
    if a>b:
        a = a - b
    else:
        b = b - a

for i in range(1,a+1):
    if a%i==0:
        print(i, end=' ')
```

```python
#W4A24

a = int(input())
sum = 0
n = 0

while sum<=a:
    n = n + 1
    sum = sum + 1/n

print(n, sum)
```

```python
#W4A25

max = int(input())
min = max

while True:
    n = int(input())
    if n == -1:
        break
    if n>max:
        max = n
    if n<min:
        min = n

print(max, min)
```

```python
#W4A29

s1 = input()
s2 = ''

for i in s1:
    s = ord(i)
    if s>96:
        s2 = s2 + chr(s-32)
    else:
        s2 = s2 + chr(s+32)

print(s2)
```

```python
#W4A31

s1 = str(input())
h, t, s = 0, 0, 0

for i in s1:
    n = ord(i)
    if n>48 and n<58:
        s = s + 1
    elif n>64 and n<91:
        h = h + 1
    else:
        t = t + 1

print(h, t, s)
```
