```python
#W5A3

def ngto(x):
    if x<2:
        return False
    
    for i in range(2,int(x**(1/2))+1):
        if x%i==0:
            return False
    
    return True
```

```python
#W5A4

def hh(x):
    sum = 0
    for i in range(1,int(x**(1/2))+1):
        if x%i==0:
            sum = sum + i
            d = x//i
            if (d != i) and (d != x):
                sum = sum + x//i
    return sum == x

n = int(input())
print(hh(n))
```

```python
#W5A5

def vt(x,a):
    for i in range(0,len(a)):
        if a[i]==x:
            return x
    return -1
```

```python
#W5A6

def gt(x):
    s = 1
    for i in range(2,x+1):
        s = s*i
    return s
```

```python
#W5A7

def mt(a,b,type):
    if type=='+':
        c = a + b
        return f'{c:.2f}'
    if type=='-':
        c = a - b
        return f'{c:.2f}'
    if type=='*':
        c = a*b
        return f'{c:.2f}'
    if type=='/':
        c = a/b
        return f'{c:.2f}'
```

```python
#W5A9

def sumx(x):
    sum = 0
    while x>0:
        sum = sum + x%10
        x = x//10
    return sum

n = int(input())
print(sumx(n))
```

```python
#W5A10

def dc(s1,s2):
    if len(s1) != len(s2):
        return False
    
    m = []
    for i in range(0,123):
        m.append(0)
    for i in range(len(s1)):
        x = ord(s1[i])
        if m[x]==0:
            m[x] = s2[i]
        else:
            if m[x] != s2[i]:
                return False
    
    return True

a = str(input())
b = str(input())

print(dc(a,b))
```

```python
#W5A12

def chia(x,y):
    a = []
    a.append(x//y)
    a.append(x-a[0]*y)
    return a

x, y = map(int, input().split())

print(chia(x,y))
```

```python
#W5A15

def tamgiac(a,b,c):
    a, b, c = a*a, b*b, c*c
    
    if (a+b==c) or (b+c==a) or (c+a==b):
        return True
    return False

a, b, c = map(int, input().split())

print(tamgiac(a,b,c))
```
